```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
df_fake = pd.read_csv('./data/Fake.csv')
df_true = pd.read_csv('./data/True.csv')
print(df_fake.head())
print(df_true.head())
```

```python
df_fake['class'] = 0
df_true['class'] = 1
```

```python
df_fake.shape, df_true.shape
```

```python
#null values
df_fake.isnull().sum()
```

```python
df_true.isnull().sum()
```

```python
#duplicate checking
duplicate = df_fake[df_fake.duplicated(subset=['text'], keep=False)]
duplicate
```

```python
df_fake = df_fake.drop_duplicates()
```

```python
df_fake.shape
```

```python
duplicate = df_true[df_true.duplicated(subset=['text'], keep=False)]
duplicate
```

```python
df_true = df_true.drop_duplicates()
df_true.shape
```

```python
#balance
from sklearn.utils import resample
df_fake = resample(df_fake, 
                            replace=False, 
                            n_samples=len(df_true), 
                            random_state=42)
```

```python
df_fake.shape, df_true.shape
```

```python
#combine datasets
df = pd.concat([df_fake, df_true], axis=0)
df.head()
```

```python
df = df.drop(['title', 'subject', 'date'], axis=1)
df.head()
```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
df.head()
```

```python
def wordopt(text):
    text = text.lower()                                     
    text = re.sub(r'\[.*?\]', ' ', text)                    
    text = re.sub(r'https?://\S+|www\.\S+', ' ', text)     
    text = re.sub(r'<.*?>+', ' ', text)                  
    text = re.sub(r'[^a-z\s]', ' ', text)               
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip() 
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    return text
    
```

```python
df['text'] = df['text'].apply(wordopt)
df.head()
```

```python
x = df['text']
y = df['class']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorization = TfidfVectorizer()
x_train = vectorization.fit_transform(x_train)
x_test = vectorization.transform(x_test)
```

```python

from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x_train, y_train)
```

```python
pred_lr = LR.predict(x_test)
LR.score(x_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(x_train, y_train)
```

```python
pred_dt = DT.predict(x_test)
DT.score(x_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GB = GradientBoostingClassifier()
GB.fit(x_train, y_train)
```

```python
pred_gb = GB.predict(x_test)
GB.score(x_test, y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(random_state=0)
RF.fit(x_train, y_train)
pred_rf = RF.predict(x_test)
RF.score(x_test, y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

```python
def output_labels(n):
    if n == 0:
        return 'Fake News'
    elif n==1:
        return 'Not a Fake News'

def testing(news):
    testing_news = {'text':[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test['text'] = new_def_test['text'].apply(wordopt)
    new_x_test = new_def_test['text']
    new_x_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_x_test)
    pred_DT = LR.predict(new_x_test)
    pred_GB = LR.predict(new_x_test)
    pred_RF = LR.predict(new_x_test)

    return print('\n\n LR prediction: {} \n DT prediction: {} \nGB prediction:{} \nRF predcition: {}'.format(output_labels(pred_LR[0]),
                                                                                                            output_labels(pred_DT[0]),
                                                                                                            output_labels(pred_GB[0]),
                                                                                                            output_labels(pred_RF[0])))
```
